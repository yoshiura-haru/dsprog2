### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
string = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"
answer = string.split()
string_list = []
for word in answer:
    string_list.append(len(word))
print(string_list)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
dictionary = {}
string = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."
positions = [1,2,5,6,8,9,16] #単語の位置

words = string.split() #スペースで分割
for i in positions: #potisionsの中身の番号を一つずつ取り出す
    word = words[i-1] #0インデックスより
    key = word[:2] #先頭2文字をkeyにする
    dictionary[key] = i #辞書に追加
print(dictionary)
```
